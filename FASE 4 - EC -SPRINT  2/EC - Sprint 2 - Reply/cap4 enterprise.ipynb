# Para melhorar a analise, fizemos suposições de um comportamento mais real dos dados para mostrar como o grupo compreende e domina as relações de cada dado recebido.

# Devido a aleatoriedade dos dados gerados não é possível tirar qualquer conclusão significativa do gráficos, pois todos tendem a se manter neutros, o que na prática não aconteceria.


```python
#imports
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#dados
df = pd.read_csv('dados_sensores.csv')
df
```

GRÁFICO DE LINHA

O gráfico mostra que a temperatura do equipamento varia de forma dinâmica ao longo do tempo o que fica bem inconclusivo mas, com dados reais suponho que teriamos um aumento gradual da temperatura durante o tempo de operação do equipamento.

```python
# Definindo a área de plotagem
plt.figure(figsize=(160, 92))
# Gráfico de linha simples
df.plot(x='Timestamp', y='Temperatura(°C)', kind='line', title='GRÁFICO DE TEMPERATURA(°C)')
# Configurando as propriedades do gráfico
plt.xlabel('TEMPO(seg)')
plt.ylabel('Temperatura(°C)')
plt.grid(True)
plt.show()
```

GRÁFICO DE DISPERSSÃO - VIBRAÇÃO E TEMPO

Caso os dados fossem reais provavelmente teriamos uma relação mais linear no tempo, possivelmente o parâmetro temperatura - que aumentaria gradativamente -influenciaria na vibração.

```python
import numpy as np
# Definindo a área de plotagem
plt.figure(figsize=(40, 24))
# Criando o gráfico de dispersão
plt.scatter(x=df['Vibracao(Hz)'], y=df['Timestamp'], color='red')
plt.title('GRÁFICO DE VIBRAÇÃO NO TEMPO')
plt.xlabel('VIBRAÇÃO(HZ)')
plt.ylabel('TEMPO(seg)')
plt.grid(True)
plt.show()
```

GÁFICO DE DISPERSÃO - VIBRAÇÃO E TEMPERATURA 

Com dados simulados temos pouca percepção de relação, do meu ponto de vista com dados reais, teriamos um acumulado de pontos progressivos, mais temperatura mais vibrção, com partes móveis mais dilatadas a tendência é de mais movimento e por consequência mais vibração.

```python
import numpy as np
# Definindo a área de plotagem
plt.figure(figsize=(40, 24))
# Criando o gráfico de dispersão
plt.scatter(x=df['Vibracao(Hz)'], y=df['Temperatura(°C)'], color='red')
plt.title('GRÁFICO DE VIBRAÇÃO NO TEMPO')
plt.xlabel('VIBRAÇÃO(HZ)')
plt.ylabel('TEMPERATURA(°C)')
plt.grid(True)
plt.show()
```

GRÁFICO BOX PLOT

O box plot fornece uma visão mais clara dos outlaiers presente em cada sensor, assim como a média de valores, como os dados gerados foram limitados, não tivemos grandes variações, na pratica valores muito fora do padrão representariam pontos de atenção cruciais.

```python
# Definindo a área de plotagem
plt.figure(figsize=(12, 4))
plt.boxplot([df['Temperatura(°C)'], df['Umidade(%)'],df['Vibracao(Hz)']]) 
plt.xticks([1, 2, 3], ['Temperatura(°C)','Umidade(%)','Vibracao(Hz)'])
# Adicionar título e rótulos aos eixos
plt.title('Box plot de Temperatura, umidade e vibração' )
plt.ylabel('Valores')
# Mostrar o gráfico
plt.show()
```

GRÁFICO DE REGRESSÃO

Para ambas as relações abaixo pode-se ver uma correlação levemente negativa, leve ao ponto de ser despresível, o que sugere que a umidade e a vibração possuem pouca influência na temperatura, possívelmente como já mencionado em minhas suposições anteriores, com dados verídicos teriamos uma correlação possitiva em ambos os casos.

```python
# Configurar o estilo
sns.set(style="whitegrid", font_scale=1.2)
plt.figure(figsize=(12, 8))

sns.regplot( data=df, x='Temperatura(°C)', y='Umidade(%)', scatter_kws={'s': 50, 'alpha': 0.6, 'edgecolor': 'w', 'linewidth': 0.5}, line_kws={'color': 'red', 'lw': 2})

# Personalizar título e rótulos
plt.title('Relação entre Temperatura e Umidade', fontsize=18, pad=20)
plt.xlabel('Temperatura (°C)', fontsize=14)
plt.ylabel('Umidade (%)', fontsize=14)

# Ajustar layout e exibir
plt.tight_layout()
plt.show()
```

```python
# Configurar o estilo
sns.set(style="whitegrid", font_scale=1.2)
plt.figure(figsize=(12, 8))

sns.regplot(data=df, x='Temperatura(°C)' , y='Vibracao(Hz)', scatter_kws={'s': 50, 'alpha': 0.6, 'edgecolor': 'w', 'linewidth': 0.5}, line_kws={'color': 'red', 'lw': 2})

# Personalizar título e rótulos
plt.title('Relação entre Temperatura e Vibração', fontsize=18, pad=20)
plt.xlabel('Temperatura (°C)', fontsize=14)
plt.ylabel('Vibracao (Hz)', fontsize=14)

# Ajustar layout e exibir
plt.tight_layout()
plt.show()
```
